```python
from qiskit import *
```

```python
from qiskit.tools.visualization import plot_bloch_multivector, plot_histogram
```

```python
circ = QuantumCircuit(1,1)
circ.h(0)
sim = Aer.get_backend('statevector_simulator')
result = execute(circ,sim).result()
stvctr = result.get_statevector(circ)
print(stvctr)
```

```python
matplotlib inline
```

```python
circ.draw(output='mpl')
```

```python
plot_bloch_multivector(stvctr)
```

```python
circ.measure([0],[0])
sim2 = Aer.get_backend('qasm_simulator')
result = execute(circ,sim2,shots = 8096).result()
counts = result.get_counts()
print(counts)
```

```python
plot_histogram(counts)
```

```python
circ = QuantumCircuit(1,1)
circ.x(0)
sim = Aer.get_backend('unitary_simulator')
result = execute(circ,sim).result()
un = result.get_unitary(circ)
print(un)
```

```python
matplotlib inline
```

```python
circ.draw(output='mpl')
```
